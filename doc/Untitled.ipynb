```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# prototype of data-generation function
x = 10 * np.random.randn(1000, 500)

F = [(np.square, None), (np.sin, None), (np.cos, None)]
funs = np.random.choice(np.arange(len(F)), 10)

y = np.zeros((1, 500))

print('Y = 0', end='')
for i in range(10):
    f,args = F[funs[i]]
    print(' + {}(X[{},],{})'.format(f.__name__, i, args), end='')
    yx = f(x[i,:], args)
    # normalize by the variance
    y += (yx - min(yx))/(max(yx) - min(yx))

x = x.T

cols = ['y']
cols.extend([ 'x' + str(i) for i in range(x.shape[1]) ])

df = pd.DataFrame(np.hstack((y.T,x)), columns = cols)
```

```python
df['tx0'] = np.cos(df['x0'])
df['tx0'] = (df['tx0'] - min(df['tx0']))/(max(df['tx0']) - min(df['tx0']))
df['tx1'] = np.sin(df['x1'])
df['tx1'] = (df['tx1'] - min(df['tx1']))/(max(df['tx1']) - min(df['tx1']))
df['tx2'] = np.square(df['x2'])
df['tx2'] = (df['tx2'] - min(df['tx2']))/(max(df['tx2']) - min(df['tx2']))
df['tx3'] = np.cos(df['x3'])
df['tx3'] = (df['tx3'] - min(df['tx3']))/(max(df['tx3']) - min(df['tx3']))
df['tx4'] = np.cos(df['x4'])
df['tx4'] = (df['tx4'] - min(df['tx4']))/(max(df['tx4']) - min(df['tx4']))
df['tx5'] = np.sin(df['x5'])
df['tx5'] = (df['tx5'] - min(df['tx5']))/(max(df['tx5']) - min(df['tx5']))
df['tx6'] = np.square(df['x6'])
df['tx6'] = (df['tx6'] - min(df['tx6']))/(max(df['tx6']) - min(df['tx6']))
df['tx7'] = np.sin(df['x7'])
df['tx7'] = (df['tx7'] - min(df['tx7']))/(max(df['tx7']) - min(df['tx7']))
df['tx8'] = np.square(df['x8'])
df['tx8'] = (df['tx8'] - min(df['tx8']))/(max(df['tx8']) - min(df['tx8']))
df['tx9'] = np.cos(df['x9'])
df['tx9'] = (df['tx9'] - min(df['tx9']))/(max(df['tx9']) - min(df['tx9']))

df['sum'] = df['tx0'] + df['tx1'] + df['tx2'] + df['tx3'] + df['tx4'] + df['tx5'] + df['tx6'] + df['tx7'] + df['tx8'] + df['tx9']
```

```python
sns.jointplot(data = df, x='sum', y='y')
```

```python
fig=plt.figure(figsize=(30, 30), dpi= 800, facecolor='w', edgecolor='k')
fig, axs = plt.subplots(nrows = 3, ncols = 3)
sns.regplot(data = df, x = 'x0', y = 'y', ax = axs[0,0])
sns.regplot(data = df, x = 'x1', y = 'y', ax = axs[0,1])
sns.regplot(data = df, x = 'x2', y = 'y', ax = axs[0,2])
sns.regplot(data = df, x = 'x3', y = 'y', ax = axs[1,0])
sns.regplot(data = df, x = 'x4', y = 'y', ax = axs[1,1])
sns.regplot(data = df, x = 'x5', y = 'y', ax = axs[1,2])
sns.regplot(data = df, x = 'x6', y = 'y', ax = axs[2,0])
sns.regplot(data = df, x = 'x7', y = 'y', ax = axs[2,1])
sns.regplot(data = df, x = 'x8', y = 'y', ax = axs[2,2])
```
