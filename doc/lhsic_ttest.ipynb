```python
import csv
from scipy import stats
import numpy as np

with open('../results/lhl_asu_scale_smalltrain_50p_delta/orlraws10P_prediction.tsv') as tsvfile:
    asu_lhsic_delta = csv.reader(tsvfile, delimiter='\t')
    asu_lhsic_delta_list = []
    for row in asu_lhsic_delta:
        if row[0] == 'localized_HSIC_lasso-K=5-S=1' and row[1]=='10':
            asu_lhsic_delta_list.append(row[3])
        print(row)
print(asu_lhsic_delta_list)
print(len(asu_lhsic_delta_list))
asu_lhsic_delta_list = np.array(asu_lhsic_delta_list, dtype=np.float)


 
with open('../results/others_asu_smalltrain_50p/orlraws10P_prediction.tsv') as tsvfile2:
    asu_hsic = csv.reader(tsvfile2, delimiter='\t')
    asu_hsic_list = []
    for row in asu_hsic:
        if row[0] == 'HSIC_lasso-B=0' and row[1]=='10':
            asu_hsic_list.append(row[3])
        print(row)
asu_hsic_list = np.array(asu_hsic_list, dtype=np.float)


asu_lhsic_delta_list
stats.ttest_ind(asu_lhsic_delta_list, asu_hsic_list, equal_var = False)
          
          
          
```
